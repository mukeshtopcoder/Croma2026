```python
# Stemming
# Stemming means reducing words to their root form
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
tokens = ['playing' , 'played' , 'players']
```

```python
new_tokens = []
for word in tokens:
    print(stemmer.stem(word))
    new_tokens.append( stemmer.stem(word) )
new_tokens
```

```python
tokens
```

```python
new_tokens = list(map( lambda val:stemmer.stem(val) , tokens ))
new_tokens
```

```python
tokens = ['Pepsi','Tesla']
new_tokens = list(map( lambda val:stemmer.stem(val) , tokens ))
new_tokens
```

```python
tokens = ['Pepsing','pepsis']
new_tokens = list(map( lambda val:stemmer.stem(val) , tokens ))
new_tokens
```

```python
tokens = ['Pepsi','Pepsis' , 'Pepsing' , "Playing" , "betters"]
for w in tokens:
    print(w,'->',stemmer.stem(w))
```
